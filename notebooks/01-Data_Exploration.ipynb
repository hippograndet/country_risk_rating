# 3 - Preprocessing

```python
from src.utils import templates
```

```python
oecd_rating_matrix = io.load_csv(config.INTERIM_DATA_DIR / 'oecd_rating_matrix_yearly.csv', index_col=0)
oecd_rating_matrix
```

```python
oecd_countries = list(oecd_rating_matrix.index)
```

```python
template_df = templates.get_template_yearly_df(oecd_countries, last_year=2023)
template_df
```

```python
dbnomics_df = io.load_csv(config.INTERIM_DATA_DIR / 'dbnomics_dataset_extracted.csv', index_col=0)
dbnomics_df
```

```python
dbnomics_df_subset = dbnomics_df[dbnomics_df.index.isin(template_df.index)]
dbnomics_df_selected = template_df.join(dbnomics_df_subset)
dbnomics_df_selected = dbnomics_df_selected[dbnomics_df_selected['ISO3_COUNTRY_CODE'].isin(oecd_countries)]
dbnomics_df_selected
```

```python
io.save_csv(dbnomics_df_selected, config.PROCESSED_DATA_DIR / 'features_dataset.csv', index=True)
```
